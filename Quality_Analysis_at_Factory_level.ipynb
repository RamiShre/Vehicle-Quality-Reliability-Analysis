```python
df
```

```python
df=df.set_index("VIN")
```

```python
df
```

```python
df=df.drop(["CAUSAL_PART_NM","TRANSACTION_ID"],axis=1)
```

```python
df
```

```python
grouped = df.groupby(['PLANT', 'GLOBAL_LABOR_CODE_DESCRIPTION']).size().reset_index(name='COUNT')
grouped
```

```python
pivot_df = grouped.pivot(index='PLANT', columns='GLOBAL_LABOR_CODE_DESCRIPTION', values='COUNT').fillna(0)
pivot_df
```

```python
pivot_df.plot(kind='bar', stacked=False, figsize=(12, 7), colormap='tab20')

plt.title('Issue Counts by Plant')
plt.xlabel('Plant')
plt.ylabel('Number of Vehicles')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Issue', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
groupbody = df.groupby(['BODY_STYLE', 'GLOBAL_LABOR_CODE_DESCRIPTION']).size().reset_index(name='COUNT')
groupbody
```

```python
pivot_body = groupbody.pivot(index='BODY_STYLE', columns='GLOBAL_LABOR_CODE_DESCRIPTION', values='COUNT').fillna(0)
pivot_body
```

```python
pivot_body.plot(kind='bar', stacked=False, figsize=(12, 7), colormap='tab20')

plt.title('Issue Counts by Body style')
plt.xlabel('Plant')
plt.ylabel('Number of Vehicles')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Issue', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
grouped_km = df.groupby(['BODY_STYLE', 'GLOBAL_LABOR_CODE_DESCRIPTION'])['KM'].mean().reset_index()
grouped_km
```

```python
pivot_km = grouped_km.pivot(index='BODY_STYLE', columns='GLOBAL_LABOR_CODE_DESCRIPTION', values='KM').fillna(0)
pivot_km
```

```python
pivot_km.plot(kind='bar', stacked=False, figsize=(12, 7), colormap='tab20')

plt.title('Bodystyle Issue')
plt.xlabel('Plant')
plt.ylabel('Average Kilometers (KM)')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Issue', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
